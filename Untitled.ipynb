```python
from collections import Counter
all_possible_iterations = list(itertools.product([0, 1], repeat=12))
all_possible_iterations_str = [''.join(map(str, state)) for state in all_possible_iterations]
all_possible_iterations_str
```

```python
probs_per_state = []
for state in all_possible_iterations:
    state = list(state)
    game = task1.Game(rule_num=110, initial_state=state)
    result = game.text_result()
    probs_per_state.append(Counter(result))
```

```python
entropy = []
for prob in probs_per_state:
    value = 0
    for iteration in all_possible_iterations_str:
        pr = prob[iteration]/100 
        if pr == 0:
            continue
        value -= float(pr * np.log2(pr))
    entropy.append(value)
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, len(entropy))
plt.plot(entropy, x)
```
